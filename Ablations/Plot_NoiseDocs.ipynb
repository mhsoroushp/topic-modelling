```python
key2add = {
    0.01: 0.02, 
    0.02: 0.02, 
    0.1: 0.02, 
    0.25: -0.02, 
    0.5: -0.04, 
    1: -0.08, 
    2: -0.20
}

def add(key):
    return key2add[key]
```

```python
def normalize_value(data, value, add):
    new_data = {}
    for key in data.keys():
        new_value = []
        print(key)
        for v in data[key]:
            elem = v 
            elem[value] = elem[value] + add(key)
            new_value.append(elem)
        new_data[key] = new_value
    return new_data
```
